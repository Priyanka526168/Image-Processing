```python
import cv2
import numpy as np

def gg(x):
    pass

# cv2.namedWindow("WIN1")
# cv2.createTrackbar("LH", "WIN1",0,255,gg)#TRACK BAR NAME,WINDOW NAME,INITIAL POSITION,MAX VALUE,FUNC
# cv2.createTrackbar("LS", "WIN1",0,255,gg)
# cv2.createTrackbar("LV", "WIN1",0,255,gg)

# cv2.createTrackbar("HH", "WIN1",255,255,gg)
# cv2.createTrackbar("HS", "WIN1",255,255,gg)
# cv2.createTrackbar("HV", "WIN1",255,255,gg)

cap = cv2.VideoCapture(0)

frame=0
back=0
forucc =cv2.VideoWriter_fourcc(*'mp4v')
out=cv2.VideoWriter('invisibility_cloak.mp4',forucc,20,(640,480))

for i in range(100):
    ret,back=cap.read()
#     if back is not None:
#         print (back.shape)
#         back=cv2.rotate(back, cv2.cv2.ROTATE_90_CLOCKWISE)
#         back=cv2.resize(back,(400,600))
if back is not None:
    cv2.imshow("BACK",back)
kernel= np.ones((5,5),np.uint8)

while(1):
    ret,frame=cap.read()
#     if frame is not None:
#         frame=cv2.rotate(frame, cv2.cv2.ROTATE_90_CLOCKWISE)
#         frame=cv2.resize(frame,(400,600))
    if frame is not None:
        hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
#     lh=cv2.getTrackbarPos("LH","WIN1")
#     ls=cv2.getTrackbarPos("LS","WIN1")
#     lv=cv2.getTrackbarPos("LV","WIN1")

#     hh=cv2.getTrackbarPos("HH","WIN1")
#     hs=cv2.getTrackbarPos("HS","WIN1")
#     hv=cv2.getTrackbarPos("HV","WIN1")
    
    lower=np.array([0,107,73])
    upper=np.array([24,255,255])
    
#     lower=np.array([lh,ls,lv])
#     upper=np.array([hh,hs,hv])
    
    mask=cv2.inRange(hsv,lower,upper)
#     mask=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
#     mask=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
    mask2=cv2.bitwise_not(mask)
    if frame is not None:
        rec1=cv2.bitwise_and(back,back,mask=mask2)
        rec2=cv2.bitwise_and(frame,frame,mask=mask)
        final=rec1+rec2

#     cv2.imshow("mask",mask)
#     cv2.imshow("mask2",mask2)
#     cv2.imshow("rec1",rec1)
#     cv2.imshow("rec2",rec2)
    cv2.imshow("output",final)
    
    if final is not None:
        out.write(final)
    
#     if frame is not None:
#         cv2.imshow("frame",frame)
#     cv2.imshow("hsv",hsv)
    if(cv2.waitKey(50)==ord("q")):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
