```python
import cv2
import numpy as np

def gg(x):
    pass

# cv2.namedWindow("WIN1")
# cv2.createTrackbar("LH", "WIN1",0,255,gg)#TRACK BAR NAME,WINDOW NAME,INITIAL POSITION,MAX VALUE,FUNC
# cv2.createTrackbar("LS", "WIN1",0,255,gg)
# cv2.createTrackbar("LV", "WIN1",0,255,gg)

# cv2.createTrackbar("HH", "WIN1",255,255,gg)
# cv2.createTrackbar("HS", "WIN1",255,255,gg)
# cv2.createTrackbar("HV", "WIN1",255,255,gg)

cap=cv2.VideoCapture("ball detection.mp4")
forucc =cv2.VideoWriter_fourcc(*'mp4v')
out=cv2.VideoWriter('ball detection_my project.mp4',forucc,20,(400,600))

while 1:
    ret,frame=cap.read()
    frame=cv2.resize(frame,(400,600))
    
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
#     lh=cv2.getTrackbarPos("LH","WIN1")
#     ls=cv2.getTrackbarPos("LS","WIN1")
#     lv=cv2.getTrackbarPos("LV","WIN1")
    
#     hh=cv2.getTrackbarPos("HH","WIN1")
#     hs=cv2.getTrackbarPos("HS","WIN1")
#     hv=cv2.getTrackbarPos("HV","WIN1")
    
    lower=np.array([131,82,155])
    upper=np.array([190,215,247])
    lower1=np.array([0,91,148])
    upper1=np.array([10,208,251])
    
    th=cv2.inRange(hsv,lower,upper)
    th1=cv2.inRange(hsv,lower1,upper1)
    
    ret,contours,heirarchy=cv2.findContours(th,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    ret,contours1,heirarchy=cv2.findContours(th1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    if(len(contours)!=0):
        max_cnt=max(contours,key=cv2.contourArea)
        x,y,w,h=cv2.boundingRect(max_cnt)
        frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,0),3)
    if(len(contours1)!=0):
        max_cnt1=max(contours1,key=cv2.contourArea)
        x1,y1,w1,h1=cv2.boundingRect(max_cnt1)
        frame=cv2.rectangle(frame,(x1,y1),(x1+w1,y1+h1),(0,0,0),3)
    value="PINK"
    cv2.putText(frame,str(value),(x-5,y-20),cv2.FONT_HERSHEY_COMPLEX,1,(147,20,255),2,cv2.LINE_AA)
    value1="ORANGE"
    cv2.putText(frame,str(value1),(x1-5,y1-20),cv2.FONT_HERSHEY_COMPLEX,1,(0,165,255),2,cv2.LINE_AA)
    
    out.write(frame)
#     cv2.imshow("WIN",th1)
    cv2.imshow("IMAGE",frame)
    if(cv2.waitKey(5)==ord("q")):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
